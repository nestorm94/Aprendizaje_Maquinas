```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_train = [[13],[15],[17],[20],[23],[25]]
y_train =[["NO"],["SI"],["SI"],["NO"],["NO"],["NO"]]
```

```python
x_test = [[14],[25],[16]]
y_test =[["SI"],["NO"],["SI"]]
```

```python
modelo = DecisionTreeClassifier(max_depth=1)
```

```python
modelo.fit(x_train,y_train)
```

```python
modelo.predict(x_train)
```

```python
plot_tree(modelo)
```

```python
modelo.predict(x_test)
```

```python
accuracy_score(y_train, modelo.predict(x_train))*100
```

```python
accuracy_score(y_test, modelo.predict(x_test))*100
```
