```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression

# Datos de ejemplo
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
OR = np.array([0, 1, 1, 1])
AND = np.array([0, 0, 0, 1])

# Entradas A y B
A = np.array([0, 0, 1, 1])
B = np.array([0, 1, 0, 1])

# Crear una figura con 2 filas y 2 columnas para los gráficos
plt.figure(figsize=(12, 10))

# Función para trazar la línea de decisión
def plot_decision_boundary(model, title):
    xx, yy = np.meshgrid(np.linspace(0, 3, 400), np.linspace(0, 3, 400))  # Cambio en el rango
    z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    z = z.reshape(xx.shape)

    plt.contourf(xx, yy, z, cmap=plt.cm.cool, alpha=0.4)
    plt.title(title)
    plt.xlabel('Entrada A')
    plt.ylabel('Entrada B')
    plt.xlim(0, 2)  # Ajuste de los límites de los ejes x
    plt.ylim(0, 2)  # Ajuste de los límites de los ejes y

# Compuerta AND
model_and = LogisticRegression()
model_and.fit(x, AND)

plt.subplot(2, 2, 2)
plt.scatter(x[:, 0], x[:, 1], c=AND, cmap=plt.cm.cool, s=200)
plot_decision_boundary(model_and, 'AND')

# Calcular la salida de la compuerta AND y mostrar en una tabla
print("Tabla de salida para la compuerta AND:")
print("    Entrada A | Entrada B     | Z        | Salida")
for i in range(4):
    A, B = x[i]
    z = model_and.coef_[0][0] * A + model_and.coef_[0][1] * B + model_and.intercept_[0]
    output = 1 / (1 + np.exp(-z))
    print(f"    {A}         |     {B}         |  {z:.4f} |  {output:.4f}")

plt.tight_layout()
plt.show()
```

```python
# Mostrar los coeficientes de la pendiente
print("Coeficientes de la pendiente del modelo AND:")
print("Coeficiente de A:", model_and.coef_[0][0])
print("Coeficiente de B:", model_and.coef_[0][1])
print("Coeficiente de Intercept:",model_and.intercept_[0])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
OR = np.array([0, 1, 1, 1])
AND = np.array([0, 0, 0, 1])
NAND = np.array([1, 0, 0, 0])
XOR = np.array([0, 1, 1, 0])

# Coeficientes e interceptos personalizados para las líneas de decisión de las compuertas
coef_custom = {
    'OR': np.array([1.0, 1.0]),
    'AND': np.array([1.0, 1.0]),
    'NAND': np.array([-1.0, -1.0]),
    'XOR': np.array([-1.0, -1.0])
}

intercept_custom = {
    'OR': -0.5,
    'AND': 1.5,
    'NAND': 0.5,
    'XOR': -0.5
}

# Crear una figura
plt.figure(figsize=(12, 10))

# Función para trazar la línea de decisión personalizada
def plot_custom_decision_boundary(coef, intercept, title):
    xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 400), np.linspace(-0.5, 1.5, 400))
    z = coef[0] * xx + coef[1] * yy + intercept
    z = (z >= 0).astype(int)

    plt.contourf(xx, yy, z, cmap=plt.cm.cool, alpha=0.4)
    plt.title(title)
    plt.xlabel('Entrada A')
    plt.ylabel('Entrada B')
    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 1.5)

# Calcular el valor "z" y la sigmoid para cada compuerta
def calculate_and_show_table(gate_name, gate_values):
    coef_gate = coef_custom[gate_name]
    intercept_gate = intercept_custom[gate_name]

    z_values = np.dot(x, coef_gate) + intercept_gate
    sigmoid_values = 1 / (1 + np.exp(-z_values))

    print(f"Tabla de valores para la compuerta {gate_name}:")
    print(" Entrada A | Entrada B | Z       | Sigmoid")
    for i in range(4):
        A, B, z, sigmoid = x[i, 0], x[i, 1], z_values[i], sigmoid_values[i]
        print(f"    {A}         |     {B}         |  {z:.4f} |  {sigmoid:.4f}")

    ax = plt.subplot(2, 2, gate_values[0])
    plt.scatter(x[:, 0], x[:, 1], c=gate_values[1], cmap=plt.cm.cool, s=200)
    plot_custom_decision_boundary(coef_gate, intercept_gate, f'{gate_name} (Línea de Decisión Personalizada)')

# Mostrar los resultados para todas las compuertas
calculate_and_show_table('OR', (1, OR))
calculate_and_show_table('AND', (2, AND))
calculate_and_show_table('NAND', (3, NAND))
calculate_and_show_table('XOR', (4, XOR))

plt.tight_layout()
plt.show()
```
