```python
def recta(x, theta1, theta0):
    return theta1 * x + theta0
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def f(x1, x2, theta1, theta0):
    return theta1 * x1 + theta0 - x2
```

```python

# Valores de theta1 y theta0 para definir la línea
theta1 = 0.5
theta0 = 2.0

 # Este valor es mayor que el valor calculado por la línea
x1_above = 4.0
x2_above = 5.0 

# Este valor es igual al valor calculado por la línea
x1_on = 2.0
x2_on = 3.0  
# Este valor es menor que el valor calculado por la línea

x1_below = 6.0
x2_below = 1.0  
```

7 - cree una función:

 $g(x_1, x_2) = sigmoid(f(x_1, x_2))$
 
 donde $f(x_1, x_2)$ es la función creada anteriormente.

```python
def g(x1, x2, theta1, theta0):
    return sigmoid(f(x1, x2, theta1, theta0))

```

```python
# Ejemplo de uso de g(x1, x2)
x1 = 2.0
x2 = 3.0
theta1 = 0.5
theta0 = 2.0
resultado_g = g(x1, x2, theta1, theta0)
print("Valor de g(x1, x2) en x1 =", x1, "y x2 =", x2, "es", resultado_g)
```

 9 - entrenar un modelo utilizando sklearn Logistic Regression. Dibujar la línea encontrada por el algoritmo. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression



# Crear un modelo de regresión logística
modelo = LogisticRegression()

# Entrenar el modelo
modelo.fit(muestras, clases)
```

```python
theta1 = modelo.coef_[0][0]
theta0 = modelo.intercept_[0]
```

```python
# Crear una gráfica para mostrar los datos de muestra y la línea
plt.scatter(muestras[:, 0], muestras[:, 1], c=clases)
x1_values = np.linspace(0, 10, 100)  # Valores de x1 para la línea
x2_values = -(theta1 * x1_values + theta0)  # Calcular x2 utilizando la ecuación de la línea

plt.plot(x1_values, x2_values, 'r-', label="Línea encontrada por el modelo")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```
