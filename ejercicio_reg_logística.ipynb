```python

# Valores de theta1 y theta0 para definir la línea
theta1 = 0.5
theta0 = 2.0

 # Este valor es mayor que el valor calculado por la línea
x1_above = 4.0
x2_above = 5.0 

# Este valor es igual al valor calculado por la línea
x1_on = 2.0
x2_on = 3.0  
# Este valor es menor que el valor calculado por la línea

x1_below = 6.0
x2_below = 1.0  
```

```python

# Calcular los valores de la línea
x = np.linspace(0, 8, 100)
line_y = recta(x, theta1, theta0)

# Graficar los puntos y la línea
plt.plot(x, line_y, 'r-', label="Línea divisoria")
plt.scatter(x1_above, x2_above, c='blue', label="Punto por encima de la línea")
plt.scatter(x1_on, x2_on, c='green', label="Punto sobre la línea")
plt.scatter(x1_below, x2_below, c='red', label="Punto por debajo de la línea")
plt.xlabel("X1")
plt.ylabel("X2")
#plt.legend()
plt.grid()
plt.show()
```

```python
# Calcular f(x1, x2) en cada punto utilizando la función definida
f_above = f(x1_above, x2_above, theta1, theta0)
f_on = f(x1_on, x2_on, theta1, theta0)
f_below = f(x1_below, x2_below, theta1, theta0)

# Imprimir los valores de f(x1, x2)
print("f(x1, x2) en el punto por encima de la línea:", f_above)
print("f(x1, x2) en el punto sobre la línea:", f_on)
print("f(x1, x2) en el punto por debajo de la línea:", f_below)
```
