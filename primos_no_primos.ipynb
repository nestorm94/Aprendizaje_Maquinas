```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
modelo = DecisionTreeClassifier()
```

```python
X=[[2],[1],[3],[8],[5],[6],[7],[9] ]
y= ["PRIMO","NO PRIMO","PRIMO","NO PRIMO","PRIMO","NO PRIMO","PRIMO","NO PRIMO"]
```

```python
modelo.fit(X, y)
```

```python
modelo.predict(X)
```

```python
plot_tree(modelo)
```

```python
X_TEST=[[11],[1],[13],[8], [17],[6], [19],[9] ]
y_TEST= ["PRIMO","NO PRIMO","PRIMO","NO PRIMO","PRIMO","NO PRIMO","PRIMO","NO PRIMO"]
```

```python
modelo.fit(X_TEST, y_TEST)
```

```python
modelo.predict(X_TEST)
```

```python
plot_tree(modelo)
```
